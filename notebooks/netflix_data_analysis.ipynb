```python
movies = data[data['type'] == 'Movie'].copy()

movies['duration_min'] = (
    movies['duration']
    .str.extract('(\d+)')[0]
    .astype('Int64')
)
# movies[movies['duration'].str.extract(r'(\d+)')[0].isna()][['title', 'duration']]
# movies['duration'].isna().sum()
int(movies['duration_min'].max())
```

```python
### most common genre
data['listed_in'].str.split(', ').explode().value_counts().index[0]
#str.split(', ') splits each genre
#explode() creates a row for each genre
# value_counts() counts repeatetion of each genre
```

```python
data['year']=pd.to_datetime(movies['date_added']).dt.year
movies=data[data['type']=='Movie'].copy()
movies_per_year=(
    movies.groupby('year')['title']
 .count()
 .reset_index(name='Number of Movies')
)
movies_per_year
```

```python
tv_shows=data[data['type']=='TV Show'].copy()
tv_shows['date_added']=tv_shows['date_added'].str.strip()
tv_shows['added_year']=pd.to_datetime(tv_shows['date_added'],errors='coerce').dt.year
tv_shows = tv_shows.dropna(subset=['added_year'])
tv_shows['added_year']=tv_shows['added_year'].astype(int)
tv_shows.groupby('added_year')['title'].count()
```
